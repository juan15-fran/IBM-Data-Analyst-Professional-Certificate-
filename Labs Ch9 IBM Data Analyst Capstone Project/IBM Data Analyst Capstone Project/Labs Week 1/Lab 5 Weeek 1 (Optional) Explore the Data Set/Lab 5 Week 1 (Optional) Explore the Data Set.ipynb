```python
import pandas as pd
```

```python
dataset_url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/LargeData/m1_survey_data.csv"
```

```python
# your code goes here
df_survey = pd.read_csv(dataset_url)
```

```python
# Identify the mean age of the survey participants
mean_age = df_survey['Age'].mean()
mean_age_rounded = round(mean_age,1)
print('The mean age of the survey participants is ', mean_age_rounded)
```
